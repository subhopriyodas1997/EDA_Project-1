```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Change data type. 
df['Amount'] = df['Amount'].astype(int)
```

```python
df.info()
```

```python
# Rename columns. 
df.rename(columns={'Marital_Status':'Relationship_Status'}, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
#Visualization
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# From graph we can see that most of the buyers are females and their purchase power is higher. 

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sales_gen_plot = sns.barplot(x='Gender',y='Amount', data = sales_gen)

for bar in sales_gen_plot.containers:
    sales_gen_plot.bar_label(bars)
```

```python
ax_new = sns.countplot(x='Age Group', data=df, hue='Gender')

for bars in ax_new.containers:
    ax_new.bar_label(bars)
```

```python
# From graph we can see that most of the buyers are between age 26-35 years and females. 

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sales_age_plot = sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sales_state_plot = sns.barplot(x='State', y='Orders', data=sales_state)

for bars in sales_state_plot.containers: 
    sales_state_plot.bar_label(bars)
```

```python
# From graph we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively. 

sales_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sales_amount_plot = sns.barplot(x='State', y='Amount', data = sales_amount)

for bars in sales_amount_plot.containers:
    sales_amount_plot.bar_label(bars)
```

```python
df.info()
```

```python
sales_relationship = sns.countplot(x='Relationship_Status', data = df)

sns.set(rc={'figure.figsize':(7,5)})
for bars in sales_relationship.containers:
    sales_relationship.bar_label(bars)
```

```python
# From grapgh we can see that most of the buyers are married women and they have high purchasing power.  

sales_status = df.groupby(['Relationship_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(7,5)})
sales_status_plot = sns.barplot(x='Relationship_Status', y='Amount', data=sales_status, hue='Gender')

for bars in sales_status_plot.containers:
    sales_status_plot.bar_label(bars)
```

```python
df.info()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_occupation = sns.countplot(x='Occupation', data = df)

for bars in sales_occupation.containers:
    sales_occupation.bar_label(bars)
```

```python
# From graph we can see that most of the buyers are working in IT, Healthcare and Aviation sector. 

sales_occu = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sales_occu_plot = sns.barplot(x='Occupation', y='Amount',data=sales_occu)

for bars in sales_occu_plot.containers:
    sales_occu_plot.bar_label(bars)
```

```python
df['Product_Category'].describe()
```

```python
sns.set(rc={'figure.figsize':(27,5)})
sales_prod = sns.countplot(x='Product_Category', data=df)

for bars in sales_prod.containers:
    sales_prod.bar_label(bars)
```

```python
# From graph we can see that most of the sold products are from Food, Clothing and Electronics categories. 

sales_categ = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(27,5)})
sales_categ_plot = sns.barplot(x='Product_Category', y='Amount', data = sales_categ)

for bars in sales_categ_plot.containers:
    sales_categ_plot.bar_label(bars)
```

```python
# Top 10 most sold products

fig1, ax1= plt.subplots(figsize=(10,3))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
